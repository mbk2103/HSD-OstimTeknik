## 5. TensorFlow/Keras ile Basit Model


```python
from tensorflow import keras
from keras import layers
#from tensorflow.keras import layers

# Bir yapay sinir ağı modeli
def build_model():
    model = keras.Sequential([
        layers.Dense(16, activation='relu', input_shape=(10,)), # Giriş katmanı.
        layers.Dense(8, activation='relu'), # Ara katman.
        layers.Dense(1, activation='sigmoid') # Çıkış katmanı. Tek nöron, sigmoid aktivasyonu (çünkü ikili sınıflandırma yapılacak — örneğin evet/hayır, 0/1).
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

print("\nTensorFlow/Keras ile Bir Model Tanımlandı.")
```

optimizer='adam': Parametre güncellemeleri için Adam optimizasyon algoritması.

loss='binary_crossentropy': İkili sınıflandırma problemleri için uygun kayıp fonksiyonu.
